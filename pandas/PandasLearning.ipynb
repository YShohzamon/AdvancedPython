# *DataFrame()* ma'lumotni "***Jadval***" ko'rnishiga olib keladi.

```python
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2],
  'year': [1997, 2004, 2010]
}

myvar = pd.DataFrame(mydataset) #Jadval ko'rinishida shakillantiradi.
print(myvar)
```

#*Series()* ma'lumotlarni "***ustun***" ko'rinishida olib keladi.

```python
a = [1, 7, 2]
myvar2 = pd.Series(a) #Ustun ko'rinishida shakillantiradi.
myvar3 = pd.Series(a, index = ["x", "y", "z"]) # qo'lda indexlash.

print(myvar2)
print("----------------------")
print(myvar3)
print("----------------------")
print(myvar2[2],"|||" , myvar3["z"]) #aynal elementni index bilan olsa bo'ladi.
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar4 = pd.Series(calories) #dict larni ham aylantirish mumkin.
print(myvar4)
```
