```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.read_csv("Salary_Data.csv")
print(df.head(8))
```

```python
print(df.info())
```

```python
new_df = df.dropna() # jadvaldan bo'sh katakli qatorlarni olib tashlaydi. (inplace = True) qilinsa asosiy fileni o'zini o'gartiradi.
print(new_df.info())
```

```python
df_salary_avg = df['Salary'].mean()
print(df_salary_avg.round())
df.fillna({"Salary": 0}, inplace=True) # bo'sh kataklarni qiymat bilan to'ldirish.
print(df.info())
```

```python
df_age_avg = df['Age'].mean()
df_age_mode = df["Age"].mode()
df_age_median = df['Age'].median()
print("Yosh o'rtachasi:", df_age_avg)
print("Yosh modasi:", df_age_mode)
print("Yosh median:", df_age_median)
```
